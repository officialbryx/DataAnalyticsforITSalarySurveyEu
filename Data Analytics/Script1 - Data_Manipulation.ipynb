```python
fill_information_values = {
    information_columns[0] : 'No Answer',
    information_columns[1] : 'No Answer',
    information_columns[2] : 'No Answer',
    information_columns[3] : 'No Answer',
    information_columns[4] : 'No Answer',
    information_columns[5] : 'No Answer',
    information_columns[6] : 'No Answer',
    information_columns[7] : 'No Answer',
    information_columns[8] : 'No Answer',
    information_columns[9] : 'No Answer',
    information_columns[10] : 'No Answer',
    information_columns[11] : 'No Answer',
    information_columns[12] : 'No Answer',
    information_columns[13] : 'No Answer',
    information_columns[14] : 'No Answer',
    information_columns[15] : 'No Answer',
    information_columns[16] : 'No Answer',
    information_columns[17] : 'No Answer',
    information_columns[18] : 'No Answer',
    information_columns[19] : 'No Answer',
}

df_2020_modified[information_columns] = df_2020_modified[information_columns].fillna(fill_information_values)
```

```python
unique_values = df_2020_modified['Have you lost your job due to the coronavirus outbreak?'].unique()
unique_values
```

```python
def standardize_job_loss_response(response):
    response = str(response).lower()  # Convert response to lowercase for easier comparison
    if 'yes' in response or 'yes but found a new one with better pay / perks' in response or 'laid off' in response or 'lost the job' in response or "i didn't but will be looking for new one because of covid" in response or 'Have been a freelancer at the beginning of year' in response:
        return 'Yes'
    elif 'no' in response or 'Laid off for a bit' in response or 'Leads and project inquiries have slowed down' in response or "didn't" in response or "did not" in response or "No, but there was a salary cut at 10% for 3 months and then at 5% for further 3 months" in response or "Lost the job but for different reason" in response:
        return 'No'
    elif 'kurzarbeitzeit for 1.5 months' in response:
        return 'No Answer'
    else:
        return 'No Answer'
    
df_2020_modified['Have you lost your job due to the coronavirus outbreak?'] = df_2020_modified['Have you lost your job due to the coronavirus outbreak?'].apply(standardize_job_loss_response)
```

```python
# Display unique values after standardization
unique_values_after = df_2020_modified['Have you lost your job due to the coronavirus outbreak?'].unique()
print(unique_values_after) #This checks unique values if they are converted to 'No', 'Yes' or 'No Answer'
```

```python
max_salary = df_2020_modified['Annual Gross Salary without Bonus + Stocks (EUR)'].max()
max_salary #Theres was a value of 999999.00 and i converted it to No Answer
```

```python
df_2020_modified['Yearly brutto salary (without bonus and stocks) in EUR'] = df_2020_modified['Yearly brutto salary (without bonus and stocks) in EUR'].replace(99999999999.0, 'No Answer')
```

```python
df_2020_modified = df_2020_modified.rename(columns={
    'Yearly brutto salary (without bonus and stocks) in EUR': 'Annual Gross Salary without Bonus + Stocks (EUR)',
    'Yearly bonus + stocks in EUR': 'Annual Bonus + Stocks (EUR)',
    'Your main technology / programming language': 'Main Programming Language',
    'Other technologies/programming languages you use often': 'Secondary Programming Languages'
})
```
